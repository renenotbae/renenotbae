# Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline     
sns.set(color_codes=True)
```

# Import Dataset


### Orders Main Dataset

```python
# importing gdrive link
sheet_url = 'https://docs.google.com/spreadsheets/d/1Anf52TDhMXArvBVjVUg8_eUv6lApU1lFLa63JHJ1QmM/edit#gid=832143546'

# transform to csv file
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

# read the file into data frame
df_order = pd.read_csv(sheet_url_trf)

# show the data frame information
df_order.info()

# show the first 10 rows of the data frame
df_order.head(10)
```

### Customer Dataset

```python
# importing gdrive link
sheet_url = 'https://docs.google.com/spreadsheets/d/1fvmBhsgnQ3snDAO0fo26-0JCDStmYtTqwjqSxFlCTtw/edit#gid=37489075'

# transform to csv file
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

# read the file into data frame
df_cust = pd.read_csv(sheet_url_trf)

# show the data frame information
df_cust.info()

# show the first 10 rows of the data frame
df_cust.head(10)
```

### Order Payment Dataset

```python
# importing gdrive link
sheet_url = 'https://docs.google.com/spreadsheets/d/1ekQGZI1VPrKI53jUeeyZrQjcCybXzDXLZ_O-x3U-WJU/edit#gid=1136279432'

# transform to csv file
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

# read the file into data frame
df_op = pd.read_csv(sheet_url_trf)

# show the data frame information
df_op.info()

# show the first 10 rows of the data frame
df_op.head(10)
```

# Combine Datasets

```python
# merge order + cust
df_merge1 = pd.merge(df_order,df_cust,left_on='customer_id',right_on='customer_id',how='inner')
df_merge1
```

```python
# merge order + cust + op
df_merge = pd.merge(df_merge1,df_op,left_on='order_id',right_on='order_id',how='inner')
df_merge
```

# Cleaning Data

```python
df_merge.info()
```

### String Grouping

```python
df_merge['payment_type'].replace('not_defined',np.nan, inplace=True)
df_merge['payment_type'].replace('boleto','voucher', inplace=True)
df_merge['payment_type'].value_counts()
```

```python
df_merge['payment_type'].value_counts()
```

```python
df_merge['order_status'].value_counts()
```

```python
df_merge['order_status'].replace('created','processing', inplace=True)
df_merge['order_status'].replace('approved','processing', inplace=True)
df_merge['order_status'].value_counts()
```

##Convert data into timestamp

```python
df_merge['order_purchase_timestamp'] = pd.to_datetime(df_merge['order_purchase_timestamp'],format='%Y-%m-%d',errors='coerce')
df_merge['order_approved_at'] = pd.to_datetime(df_merge['order_approved_at'],format='%Y-%m-%d',errors='coerce')
df_merge['order_delivered_carrier_date'] = pd.to_datetime(df_merge['order_delivered_carrier_date'],format='%Y-%m-%d',errors='coerce')
df_merge['order_delivered_customer_date'] = pd.to_datetime(df_merge['order_delivered_customer_date'],format='%Y-%m-%d',errors='coerce')
df_merge['order_estimated_delivery_date'] = pd.to_datetime(df_merge['order_estimated_delivery_date'],format='%Y-%m-%d',errors='coerce')
```

##Remove irrelevant and null value


```python
#remove irrelevant value
df_merge.drop(columns=['customer_unique_id','customer_zip_code_prefix'],inplace=True)
df_merge.info()
```

```python
# search null data in orders table
df_merge.isna().sum()
```

```python
df_dropna = df_merge.copy()
```

```python
df_dropna[df_dropna.isnull().any(axis=1)]
```

```python
df_dropna.dropna(axis=0,inplace=True)
df_dropna.info()
```

## Check & Remove Duplicate Values

```python
df_mergec= df_dropna.copy()
```

```python
df_mergec.duplicated().sum()
```

```python
df_mergec2 = df_mergec.drop_duplicates()
df_mergec2.duplicated().sum()
```

##Handling Outlier

```python
Q1 = df_mergec['payment_value'].quantile(.25)
Q3 = df_mergec['payment_value'].quantile(.75)
IQR = Q3 - Q1

UIF = Q3 + (1.5*IQR)
LIF = Q1 - (1.5*IQR)

print('IQR =',IQR)
print('UIF =',UIF)
print('LIF =',LIF)
```

```python
df = df_mergec[(df_mergec['payment_value']<=UIF) & (df_mergec['payment_value']>=LIF)]
df
```

```python
df.info()
```

```python
df.boxplot(column=['payment_value'],fontsize=12,
           rot=0,grid=False,figsize=(10,10),vert=False)
```

##Payment Type

```python
df_op1= df.groupby('payment_type')['order_id'].agg('count').reset_index()
df_op1
```

```python
def label_function(val):
    return f'{val / 100 * len(df):.0f}\n{val:.0f}%'

fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 5))
explode = (0.1, 0, 0)

df.groupby('payment_type').size().plot(
    kind='pie'
    , autopct=label_function
    , textprops={'fontsize': 12}
    ,colors=['#CAD824', '#EDF1D1', '#EEEEED'], ax=ax1,
    explode=explode
    )
plt.tight_layout()
plt.show()
```

Dominant payment method used is credit card, followed by voucher in the second place.

##Define Busiest Day

```python
df['order_purchase_date'] = df['order_purchase_timestamp'].dt.date
df['order_purchase_week'] = df['order_purchase_timestamp'].dt.week
df['order_purchase_month'] = df['order_purchase_timestamp'].dt.month
df['order_purchase_year'] = df['order_purchase_timestamp'].dt.year
df['order_purchase_days']=df['order_purchase_timestamp'].dt.day_name()
df['order_purchase_monthyear']= df['order_purchase_timestamp'].dt.to_period('M')
df.head(10)
```

```python
countdays=df.groupby('order_purchase_days')['order_id'].agg('count').reset_index()
countdays
```

```python
countdays.plot.bar('order_purchase_days', 'order_id', color=["#E2E8BF", "#B1C538", "#E2E8BF", "#E2E8BF",'#E2E8BF','#B1C538','#B1C538'])
plt.title('Busiest Day to Get Order')
plt.xlabel('Day')
plt.ylabel('#orders')
plt.show()
```

The busiest days are on: Monday, Tuesday, and Wednesday
